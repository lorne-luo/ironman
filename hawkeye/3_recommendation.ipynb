```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/opt/hawkeye/pic/result.csv'
df = pd.read_csv(path,
                 usecols=[
                     'code', 'last_date', 'start price', 'sim_mean', 'sim_diff',
                     'VaR 99%', 'VaR 99% Percent', 'volume_mean',
                     'return_mean', 'return_sigma'
                 ])
df.head()
df['sim_diff'] = round(df['sim_diff'] / df['start price'] * 100, 3)
print(len(df))

print(df['start price'].quantile(0.7))
print(df['sim_diff'].quantile(0.9))

volume_threshold = df['volume_mean'].quantile(0.2)
df = df[(df['volume_mean'] > volume_threshold) & (df['start price'] > 0.15)]
print(len(df))
df.sort_values(by='sim_diff', ascending=False).head(20)
```

```python
# high mean return 
df.sort_values(by='sim_diff', ascending=False).head(10)
```

```python
# high price
df[df['start price']>1].sort_values(by='sim_diff', ascending=False).head(10)
```

```python
# highest price
df[df['sim_diff']>0.8].sort_values(by='start price', ascending=False).head(10)
```

```python
# low risk
var_threshold = df['VaR 99% Percent'].quantile(0.3)
print(var_threshold)
df[df['VaR 99% Percent'] < var_threshold].sort_values(by='sim_diff',
                                                      ascending=False).head(10)
```

```python
# high volume
df.sort_values(by='volume_mean', ascending=False).head(10)
```

```python
# plt.figure(figsize=(16,8))
df['volume_mean'].quantile(0.2)
```

```python
# 2019-05-17 recommendations
selection = [
    'AVH',  # high return & high risk
    'ANO',  # high return & high risk
    'AGH',  # high return & meduim risk
    'Z1P',  # medium price & risk
    'MFG',  # high price & low risk
    'AD8',  # low risk
    'FMG',  # high volume & low risk
]
df[df['code'].isin(selection)].sort_values(by='sim_diff', ascending=False).head(10)
```
